```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

data = pd.read_csv('https://confrecordings.ams3.digitaloceanspaces.com/pubg-weapon-stats.csv')
print(data.head())
print(data.shape)
print(data.isnull().sum())
print(data.dtypes)
print(data.columns)



print(data.dropna())
corr_data = data.corr()
print(corr_data)
```
